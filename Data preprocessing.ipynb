```python
def neighbour_duplicates(data):
  while (data['Operation code'].diff() == 0).astype(int).sum() != 0:

    data['diff'] = data['Operation code'].diff()
    duplicate = data[data['diff']==0].index -1

    for i in duplicate:

      data['Time, h (in grains)'][i] =data['Time, h (in grains)'][i+1] + data['Time, h (in grains)'][i]
    data = data.drop(index=duplicate+1)

  return data

#(data['Код операции'].diff() == 0).astype(int).sum()


```
